## Exploratory Data Analysis

### gender

```python
 ax = sns.countplot(x = 'Gender', data = data)
```

```python
gen_sales = data.groupby(['Gender'],as_index = False) ['Amount'].sum().sort_values(by= 'Amount', ascending = False)

sns.barplot( x= 'Gender',y = 'Amount', data = gen_sales)
            
                       
```

###### from over graph we can see the most buyer are female and even the parchasing power in female more then male

### AGE

```python
ax =sns.countplot( data= data ,x = "Age Group" , hue= "Gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# total amount vs age 

age_amt = data.groupby(['Age Group'],as_index = False) ['Amount'].sum().sort_values(by = 'Amount',ascending = False)
sns.barplot (x='Age Group',y= 'Amount' ,data= age_amt )                 
```

###### we can see the most buyer of the female age group between 26-35 years

### oders

```python
# total no of oder in top 10 states
state_sale =  data.groupby(['State'],as_index = False) ['Orders'].sum().sort_values(by = 'Orders',ascending = False).head(10)
sns.set(rc= {'figure.figsize':(17,6)})
sns.barplot(x='State',y='Orders',data=state_sale)
```

##### from the graph we can see that most oders from utter paardesh

### Marital_Status

```python
ax=sns.countplot(data= data ,x='Marital_Status')
sns.set(rc= {'figure.figsize':(5,1)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = data.groupby(['Marital_Status','Gender'],as_index=False) ['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc= {'figure.figsize':(8,4)})
sns.barplot(x = 'Marital_Status',y ='Amount', hue='Gender' ,data=sales_state)
```

##### from above graph we can see the most buyer are married(women) they have high parchasing power

## Occupation

```python
ax=sns.countplot(data= data ,x='Occupation')
sns.set(rc= {'figure.figsize':(20,3)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Occupation=data.groupby(['Occupation'],as_index=False) ['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc= {'figure.figsize':(10,4)})
sns.barplot(x = 'Occupation',y ='Amount', data=Occupation)
```

##### above graph we can see the most of the buyer are IT SECTOR,HEALTHCARE,AVIATION

## Product_Category

```python
ax=sns.countplot(data= data ,x='Product_Category')
sns.set(rc= {'figure.figsize':(20,3)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Product_Category=data.groupby(['Product_Category'],as_index=False) ['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc= {'figure.figsize':(20,7)})
sns.barplot(x = 'Product_Category',y ='Amount', data=Product_Category)
```

##### above graph we can see the most of sold product are food clothing & apparel and eletronic & gadgets
